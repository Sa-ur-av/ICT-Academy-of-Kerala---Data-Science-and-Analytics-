### 5. Meaningful visualizations using the dataset

```python
column_name = {'SL': 'Sepal_Length','SW': 'Sepal_Width','PL': 'Petal_Length','PW': 'Petal_Width'}
df = df.rename(columns=column_name)
```

##### a. Boxplot

```python
sns.boxplot(df,palette='magma')
plt.show()
```

##### b. Heatmap

```python
correlation = df.corr(numeric_only=True)
```

```python
sns.heatmap(correlation, annot=True, cmap='magma')
plt.title('Correlation Heatmap')
```

##### c. Violinplot

```python
sns.violinplot(x='Classification', y='Sepal_Length', data=df,palette='magma')
plt.title('Sepal length of different classifications')
plt.show()
```

##### d. Scatterplot

```python
sns.scatterplot(x='Sepal_Length', y='Sepal_Width', hue='Classification', data=df, palette='magma')
plt.title('Sepal Length vs Sepal Width')
plt.legend(loc=(1.02,0))
plt.show()
```
